## Step 1: Import Libraries and Read Dataset

```python
import tensorflow as tf
import requests
import io
import matplotlib.pyplot as plt
import pandas as pd
```

## Step 2: Data Preparation

```python

dataset_url = 'https://raw.githubusercontent.com/Apress/artificial-neural-networks-with-tensorflow-2/main/Ch05/winequality-white.csv'
dataset = pd.read_csv(dataset_url, sep=';')
```

```python
# Display last few records
dataset.tail()
```

From the above , we can clearly see that there are 4897 records in the database. Each record contains 12 fields, the last one is the quality of wine that we will be using as the label.

## Step 3: Data Preprocessing

```python
# Now we will extract features and labels with the following
x = dataset.drop('quality', axis=1)
y = dataset['quality']
```

```python
# Create training, validation, testing datasets
from sklearn.model_selection import train_test_split
x_train_1, x_test, y_train_1, y_test = train_test_split (x, y, test_size=0.15, random_state=0)
x_train, x_val, y_train, y_val = train_test_split(x_train_1, y_train_1, test_size=0.05, random_state=0)
```

We have reserved 15% of the data for testing. The training data is split into 95:5 ratio: 95% training, 5% validation.

```python
# Scaling the data 
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
x_train_new = sc_x.fit_transform(x_train)
```

## Step 4: Checking preprocessing results

```python

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(20, 10))

ax1.scatter(x_train.index,
                x_train['fixed acidity'],
                color= 'blue',
                label='raw',
                alpha=0.4,
                marker='^'
                )

ax2.scatter(x_train.index,
                x_train_new[: , 1],
                color = 'green',
                label='adjusted',
                alpha=0.4,
                marker='s'
                )

ax1.set_title('Training dataset')
ax2.set_title('Standardized training dataset')

for ax in (ax1, ax2):
    ax.set_xlabel('index')
    ax.set_ylabel('fixed acidity')
    ax.legend(loc='upper right')
    ax.grid()

plt.tight_layout()

plt.show()
```

The plot on the left hand side above shows the raw data , where you can see that the fixed acidity field values range from 0 to 14 with the mean around 7. The plot on Right hand shows the same data after the transformation.  Mean is about zero with the data variation between -2 and 2 indicating std deviation of +/-1. 

```python

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(20, 10))

for l, c, m in zip(range(0, 2), ('blue', 'red'), ('^', 's')):
    ax1.scatter(x_train['residual sugar'],
                x_train['total sulfur dioxide'],
                color= c,
                label='class %s' % l,
                alpha=0.4,
                marker=m
                )

for l, c, m in zip(range(0, 2), ('blue', 'green'), ('^', 's')):
    ax2.scatter(x_train_new[: , 3],
                x_train_new[: , 6],
                color = c,
                label='class %s' % l,
                alpha=0.4,
                marker=m
                )

ax1.set_title('Training dataset')
ax2.set_title('Standardized training dataset')

for ax in (ax1, ax2):
    ax.set_xlabel('residual sugar')
    ax.set_ylabel('total sulfur dioxide')
    ax.legend(loc='upper right')
    ax.grid()

plt.tight_layout()

plt.show()
```

This time we plotted the total sulphur dioixide on the y axis and residual sugar on the x axis.

From the above we can see that Residual sugar value vary between 0 and 30 while the total sulphur dioxide value has a large variation between 0 and 400.

After transformation as seen in the right hand side of the above figure, both the fields show a mean value of 0 and even distribution between -2 and + 2 for the rest of data points.

This type of data transformation helps neural netwroks learn better. So we do these transformations on all the features.

```python
# Now to transform the testing and validation datasets, we simply call the trnasform method and not the fit_transform method.
x_test_new = sc_x.transform (x_test)
x_val_new = sc_x.transform(x_val)
```

## Step 5: Model Building

```python
# Metrics Vidualization function


import matplotlib.pyplot as plt
epoch = 30 
def plot_learningCurve(history):
  # Plot training & validation accuracy values
  epoch_range = range(1, epoch+1)
  #plotting the mae vs epoch of training set
  plt.plot(epoch_range, history.history['mae'])
  #plotting the val_mae vs epoch of the validation dataset.
  plt.plot(epoch_range, history.history['val_mae'])
  plt.ylim([0, 2])
  plt.title('Model mae')
  plt.ylabel('mae')
  plt.xlabel('Epoch')
  plt.legend(['Train', 'Val'], loc='upper right')
  plt.show()

  print("--------------------------------------------------------")

  # Plot training & validation loss values
  plt.plot(epoch_range, history.history['loss'])
  plt.plot(epoch_range, history.history['val_loss'])
  plt.ylim([0, 4])
  plt.title('Model loss')
  plt.ylabel('Loss')
  plt.xlabel('Epoch')
  plt.legend(['Train', 'Val'], loc='upper right')
  plt.show()
```

## Small Model

- In this small model we will build model with only 1 hidden layer



```python
# Design consist of setting 16 neurons in the hidden layer with ReLU activation on each neuron and the output is a single layer that gives out the wine quality


small_model = tf.keras.Sequential([
                            tf.keras.layers.Dense(16 , 
                                                  activation = 'relu' , 
                                                  input_shape = (11 , )),
                            tf.keras.layers.Dense(1)
])
small_model.compile(optimizer = 'adam' , loss = 'mse' , metrics = ['mae'])
history_small = small_model.fit(x_train_new, y_train , 
                                batch_size = 32, 
                                epochs = 30, verbose = 0, 
                                validation_data = (x_val_new , y_val))
```

## Small Model Evaluation Metrics

```python
plot_learningCurve(history_small)
```

## Evaluate the model's performance on the test data

```python

s_test_loss , s_test_mae = small_model.evaluate(x_test_new , y_test , batch_size=32 , verbose=1)
print("small model test_loss : {}".format(s_test_loss))
print("small model test_mae : {} ".format(s_test_mae))
```

## Evaluate the model's performance on the Unseen Data

### Create such data here we are picking up the data point from the test data with id equal to 2125.

```python
import numpy as np
unseen_data = np.array([[6.0 , 0.28 , 0.22 , 12.15 , 0.048 , 42.0 , 163.0 , 0.99570 , 3.20 , 0.46 , 10.1]])
```

```python

y_small = small_model.predict(sc_x.transform(unseen_data))
print ("Wine quality on unseen data (small model): ", y_small[0][0])
```

Wine quality on unseen data with small model = 5.5537

```python
small_model.summary()
```

## Medium Model

- In this medium model we will increase the number of hidden layers from 1 to 3.
- We will also increase number of neurons in each of these layers to 64
- We will continue using Adam optimizerand let mse as our loss function

```python

medium_model = tf.keras.Sequential([
                               tf.keras.layers.Dense(64 , activation = 'relu' , 
                                                     input_shape = (11, )),
                               tf.keras.layers.Dense(64 , activation = 'relu'),
                               tf.keras.layers.Dense(64 , activation = 'relu'),
                               tf.keras.layers.Dense(1)
])
medium_model.compile(loss = 'mse' , optimizer= 'adam' , metrics = ['mae'])
history_medium = medium_model.fit(x_train_new , y_train , batch_size=32, 
                                  epochs = 30, verbose = 1 , 
                                  validation_data= (x_val_new , y_val))
```

In the above medium model, we have added more layers to check whether adding more layers to the network would help in getting better accuracy.

## Medium Model Evaluation

```python
plot_learningCurve(history_medium)
```

- As we can observe from the above plot, the training saturates after about 5 epochs.
- If we compare this against Small model , there we didnt see the saturation untill 20 epochs.
- In the medium_model, we also observe some overfitting. Divergence in the 2 curves shown in each of the plot above showing some overfitting

## Evaluate the model's performance on the test data

```python
m_test_loss , m_test_mae = medium_model.evaluate(x_test_new , y_test , batch_size=32 , verbose = 1)
print("medium model test_loss : {}".format(m_test_loss))
print("medium model test_mae : {}".format(m_test_mae))
```

In comparision to small_model, both test_loss and test_mae have decreased in value.

## Evaluate the model's performance on the Unseen Data

```python
y_medium = medium_model.predict(sc_x.transform(unseen_data))
print ("Wine quality on unseen data (medium model): ", y_medium[0][0])
```

## Large Model

- In this medium model we will add 2 more hidden layers to our medium model making it total 4 hidden layers
- We will also increase number of neurons in each of these layers to 128
- Adding the layers and number of neurons increases the number of parameters to be tuned
- We will continue using Adam, mse, mae as before 

```python
large_model = tf.keras.Sequential([
                               tf.keras.layers.Dense(128 , activation = 'relu' , 
                                                     input_shape = (11, )),
                               tf.keras.layers.Dense(128 , activation = 'relu'),
                               tf.keras.layers.Dense(128 , activation = 'relu'),
                               tf.keras.layers.Dense(128 , activation = 'relu'),
                               tf.keras.layers.Dense(1)
])

large_model.compile(loss = 'mse' , optimizer= 'adam' , metrics = ['mae'])


history_large = large_model.fit(x_train_new , y_train , batch_size=32, epochs = 30,verbose = 1 , validation_data= (x_val_new , y_val))
     
```

## Large Model Evaluation

```python
plot_learningCurve(history_large)
```

From the above, we can observe that saturation in training is achieved much earlier after about 3 epochs. Hence this is a case of overfitting.

## Evaluate the model's performance on the test data

```python
l_test_loss , l_test_mae = large_model.evaluate(x_test_new , y_test , batch_size = 32 , verbose = 1)
print("large model test_loss : {}".format(l_test_loss))
print("large model test_mae : {}".format(l_test_mae))
```

The loss is now 0.55 and mae is 0.56

## Evaluate the model's performance on the Unseen Data

```python
y_large = large_model.predict(sc_x.transform(np.array([[6.0 , 0.28 , 0.22 , 12.15 , 0.048 , 42.0 , 163.0 , 0.99570 , 3.20 , 0.46 , 10.1]])))
print ("Wine quality on unseen data (large model): ", y_large[0][0])
```

## Reducing Overfitting in case of Large Model

### Here we will add dropout layers to this large model to check if it reduces overfitting.

```python
large_model_overfit = tf.keras.Sequential([
                               tf.keras.layers.Dense(128 , activation = 'relu' , 
                                                     input_shape = (11, )),
                               tf.keras.layers.Dropout(0.4),
                               tf.keras.layers.Dense(128 , activation = 'relu'),
                               tf.keras.layers.Dropout(0.3),
                               tf.keras.layers.Dense(128 , activation = 'relu'),
                               tf.keras.layers.Dropout(0.2),
                               tf.keras.layers.Dense(128 , activation = 'relu'),
                               tf.keras.layers.Dense(1)
])
large_model_overfit.compile(loss = 'mse' , optimizer= 'adam' , metrics = ['mae'])
history_large_overfit = large_model_overfit.fit(x_train_new , y_train , batch_size=32, epochs = 30,verbose = 0 , validation_data= (x_val_new , y_val))
plot_learningCurve(history_large_overfit)
```

- From the above we can clearly see that adding dropout layers eliminates or at least redices the overfitting.

- The droput layers are usually added to the large networks as they have enough number of neurons available in each layer from which we can afford to drop out the outputs of a few neurons.

## Training Optimization in case of Small Model( Using RMSprop Optimization)

```python
model_small = tf.keras.Sequential([
                tf.keras.layers.Dense(16 , 
                    activation = 'relu' , 
                    input_shape = (11 , )),              
              tf.keras.layers.Dense(1)
])
optimizer = tf.keras.optimizers.RMSprop(0.001)
model_small.compile(loss = 'mse' , optimizer = optimizer , metrics = ['mae'])
history_small_overfit = model_small.fit(x_train_new , y_train , batch_size=32, 
                        epochs = 30, verbose = 0 , 
                        validation_data= (x_val_new , y_val))
plot_learningCurve(history_small_overfit)
```

- Above graphs are clearly showing that RMSprop is reducing the number of epochs considerably.

- This will reduce the training time which is crucial in the case of huge datasets

- Another way of reducing training time is to use the EarlyStopping feature of Tensorflow.
